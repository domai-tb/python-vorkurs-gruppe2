### Logikoperatoren


Häufig möchte man mehrere Variablen miteinander vergleichen. 
Um beispielsweise zu überprüfen, ob eine Zahl größer als zwei weitere Zahlen ist, braucht man weitere Logikoperatoren:

* `and`
vergleicht zwei Ausdrücke miteinander und liefert den Wahrheitswert `True`, wenn **alle** Ausdrücke wahr sind.

* `or`
vergleicht zwei Ausdrücke miteinander und liefert den Wahrheitswert `True`, wenn **mindestens ein** Ausdruck wahr ist.

* `not` invertiert einen Ausdruck, wenn er also falsch ist (`False`), ist er jetzt nun wahr (`True`) und umgekehrt.











````python
x = 7
y = 8
z = 9

# Aufgabe: 
#1) checke, ob y größer als x und z ist
#2) checke, ob y größer als x oder z ist
#3) checke, ob y größer als x und z. negiere den Ausdruck. 

print(y > x and y > z)  # (True and False) = False 
print(y > x or y > z)  #  (True or False) = True
print(not(y > x and y > z)) # (True and False) = not(False)   
````

# `if`-"else" -Bedingungen
`if`-Bedingungen dienen zur bedingten Ausführung von Codeabschnitten. Ist die Bedingung wahr, wird ein der Code ausgeführt.

`if`-Blöcke werden mit dem Schlüsselwort `if` eingeleitet.
Ist die Bedingung nicht wahr, so kann man mithilfe des Schlüsselworts `else`
ausdrücken, was ansonsten passieren sollen. Dies entspricht "Wenn... dann... ansonsten..."-Bedingungen aus der natürlichen Sprache.

Beispiel: Wenn morgen die Sonne scheint, gehe ich ins Kino (`if`), ansonsten bleibe ich zuhause (`else`).

Der `else`-Block ist optional. Das heißt, eine `if`-Bedingung funktioniert auch ohne einen `else`-Block. 


````python
x = 9

if x < 11: 
  print("Die Zahl ist kleiner als 11")


````

Eine `while`-Schleife führt ihre Anweisungen aus, solange ihre Bedingung erfüllt ist.

# `try` und `except`

Mit dem `try`-Block kann man Fehler abfangen, sobald sie auftreten. Bei einem Fehler wird der Inhalt des `except`-Blocks ausgeführt.

````python
x = 10
try:
  number = int(input("Enter a number: "))
  erg = x/number
  print(erg)
except ValueError:
  print("invalid number")
except ZeroDivisionError:
  print("divided by zero")
````

# `break`, `continue` und `pass`

* `break`: Die `break`-Anweisung beendet die aktuelle Schleife und setzt die Ausführung nach der Schleife Anweisung fort.
* `continue`: Die `continue`-Anweisung bricht den aktuellen Schleifendurchlauf ab und beginnt den nächsten.
* `pass`: Die `pass`-Anweisung ist ein Platzhalter für leere Codeblöcke in z.B. Schleifen oder Bedingungen.





---

`break`

````python
for i in "Python":
  if i == "t":  # t == t
    break
  print(i)

print("The end")
````



---


`continue`

````python
for i in "Python":
  if i == "t":  # t == t
    continue
  print(i)

print("The end")
````



---
`pass`


````python
for i in "Python":
  if i == "t":  # t == t
    pass
  print(i)

print("The end")
````

## Aufgabe 6


Erstelle eine Passwort-Abfrage, bei der das Passwort nur aus Ziffern besteht.

Bei drei Fehleingaben erscheint die Botschaft ("three times wrong"). Wenn der Nutzer Zeichenketten eingibt, zählt dies nicht als Fehlversuch. Wenn der Nutzer das Passwort richtig eingibt, erscheint die Botschaft "nice!".

````python
schloss = True
password = 123
count = 0

while schloss:
  try:
    eingabe = int(input("Passwort: "))
    if eingabe == password:
      print("nice")
      schloss = False
    else:
      count = count + 1
      if count == 3:
        print("three times wrong")
        break
  except ValueError:
    print("keine Zahl")
````

# Funktionen

Mit Funktionen können Anweisungen gruppiert und wiederverwendet werden. Funktionen können Werte als Parameter erhalten und einen Wert zurückgeben.

````python
def taschenrechner(x,y):
  add = x+y
  sub = x-y
  mul = x*y
  div = x/y

  return add, sub, mul, div

print(taschenrechner(10,2))
````
