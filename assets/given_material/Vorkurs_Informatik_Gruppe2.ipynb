### Logikoperatoren


Hufig mchte man mehrere Variablen miteinander vergleichen. 
Um beispielsweise zu berprfen, ob eine Zahl grer als zwei weitere Zahlen ist, braucht man weitere Logikoperatoren:

* `and`
vergleicht zwei Ausdrcke miteinander und liefert den Wahrheitswert `True`, wenn **alle** Ausdrcke wahr sind.

* `or`
vergleicht zwei Ausdrcke miteinander und liefert den Wahrheitswert `True`, wenn **mindestens ein** Ausdruck wahr ist.

* `not` invertiert einen Ausdruck, wenn er also falsch ist (`False`), ist er jetzt nun wahr (`True`) und umgekehrt.











````python
print(not True)
print(True and True)
print(True or True)
print((True or not False) and not (True and False))
````

# `if`-"else" -Bedingungen
`if`-Bedingungen dienen zur bedingten Ausfhrung von Codeabschnitten. Ist die Bedingung wahr, wird ein der Code ausgefhrt.

`if`-Blcke werden mit dem Schlsselwort `if` eingeleitet.
Ist die Bedingung nicht wahr, so kann man mithilfe des Schlsselworts `else`
ausdrcken, was ansonsten passieren sollen. Dies entspricht "Wenn... dann... ansonsten..."-Bedingungen aus der natrlichen Sprache.

Beispiel: Wenn morgen die Sonne scheint, gehe ich ins Kino (`if`), ansonsten bleibe ich zuhause (`else`).

Der `else`-Block ist optional. Das heit, eine `if`-Bedingung funktioniert auch ohne einen `else`-Block. 


````python
zahl = int(input("Gebe eine Zahl ein: "))

if zahl < 100:
  print(f"{zahl} kleiner als 100")
else:
  print(f"{zahl} ist grer gleich 100")
````



---
**Mehrere If-Anweisungen**

*   36,5C - 38,0C: Normal-Temperatur
*   38,1C - 38,5C: Leichtes Fieber
*   38,6C - 39,0C: Miges Fieber
*   39,1C - 39,9C: Hohes Fieber
*   40,0C - 42,0C: Sehr hohes Fieber


````python
grad = 38.55

if grad <= 38.0 and grad >= 36.5:
  print("Normal-Temperatur")
elif grad >= 38.1 and grad <= 38.5:
  print("Leichtes Fieber")
elif grad >= 38.6 and grad <= 39.0:
  print("Miges Fieber")
elif grad >= 39.1 and grad <= 39.9:
  print("Hohes Fieber")
elif grad >= 40.0:
  print("Sehr hohes Fieber")
else:
  print("Unterkhlt")
````

# `break`, `continue` und `pass`

* `break`: Die `break`-Anweisung beendet die aktuelle Schleife und setzt die Ausfhrung nach der Schleife Anweisung fort.
* `continue`: Die `continue`-Anweisung bricht den aktuellen Schleifendurchlauf ab und beginnt den nchsten.
* `pass`: Die `pass`-Anweisung ist ein Platzhalter fr leere Codeblcke in z.B. Schleifen oder Bedingungen.





---

`break`

````python
for i in "Python":
  if i == "t":
    break
  print(i)
````



---


`continue`

````python
for i in "Python":
  if i == "t":
    continue
  print(i)
````



---

````python
for i in "Python":
  if i == "t":
    pass
  print(i)
````

Erstelle eine Funktion, die dir errechnet wie viele Mnzen du fr eine Bestimmte Eingabe von Cents erhltst.



>Beispiel: 200 Cent = eine 2 Mnze

>`Anzahl der Mnzen: 1`



>Beispiel: 250 Cent = eine 2 Mnze und eine 50 Cent Mnze

>` Anzahl der Mnzen: 2`

>Beispiel: 75 Cent = eine 50 Cent Mnze, eine 20 Cent Mnze, eine 5 Cent Mnze                                                                                                                                    
>`Anzahl der Mnzen: 3`




````python
def zhleMnzen(anzahl):
  mnzen = [200, 100, 50, 20, 10, 5, 1]
  anzahlmnzen = 0

  for mnze in mnzen:
    anzahlmnzen = anzahlmnzen + anzahl // mnze
    anzahl = anzahl % mnze

  return anzahlmnzen


cents = int(input("Anzahl der Cents: "))
print(f"Anzahl der Mnzen betrgt: {zhleMnzen(cents)}")
````

# Klassen

Python ist eine objekt-orientierte Programmiersprache. Dies bedeutet dass in Pythen nahezu alles ein Objekt ist. Beispiele sind `str`, `int` oder `list` welche Objekt-Typen (= Klasse) sind. Wenn wir ein neues Objekt erzeugen bspw. `name = "Tim"`, dann ist die Variable `name` eine Instanz der Klasse `str`.

Jedes Objekt besteht aus Attributen (Eigenschaften) und Methoden (Funktionen). Fr den Objekt-Typ des `dict` kennen wir beispielsweise die Methode `keys()`. 

Python erlaubt es uns jedoch auch eigene Klassen zu definieren, wozu es das Keyword `class` gibt.

````python
# Definition der Klasse
class Student():
  name = "Max Mustermann"
  alter = 21
  kurse = ["Vorkurs Informatik", "Informatik 1", "Mathematik 1"]

# Erzeuge ein Objekt (Instanz) der Klasse
schler = Student()

# Hinweis: print(*liste) <-> print(liste[0], liste[1], ..., liste[-1])  
print(f"Unser Schler heist {schler.name}, ist {schler.alter} Jahre alt und"
      f" in folgende Kurse eingeschrieben:", *schler.kurse, sep="\n - ", end="\n\n")
````
