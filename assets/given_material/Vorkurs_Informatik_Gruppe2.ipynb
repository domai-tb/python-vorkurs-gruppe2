### Logikoperatoren


Häufig möchte man mehrere Variablen miteinander vergleichen. 
Um beispielsweise zu überprüfen, ob eine Zahl größer als zwei weitere Zahlen ist, braucht man weitere Logikoperatoren:

* `and`
vergleicht zwei Ausdrücke miteinander und liefert den Wahrheitswert `True`, wenn **alle** Ausdrücke wahr sind.

* `or`
vergleicht zwei Ausdrücke miteinander und liefert den Wahrheitswert `True`, wenn **mindestens ein** Ausdruck wahr ist.

* `not` invertiert einen Ausdruck, wenn er also falsch ist (`False`), ist er jetzt nun wahr (`True`) und umgekehrt.











````python
print(not True)
print(True and True)
print(True or True)
print((True or not False) and not (True and False))
````

# `if`-"else" -Bedingungen
`if`-Bedingungen dienen zur bedingten Ausführung von Codeabschnitten. Ist die Bedingung wahr, wird ein der Code ausgeführt.

`if`-Blöcke werden mit dem Schlüsselwort `if` eingeleitet.
Ist die Bedingung nicht wahr, so kann man mithilfe des Schlüsselworts `else`
ausdrücken, was ansonsten passieren sollen. Dies entspricht "Wenn... dann... ansonsten..."-Bedingungen aus der natürlichen Sprache.

Beispiel: Wenn morgen die Sonne scheint, gehe ich ins Kino (`if`), ansonsten bleibe ich zuhause (`else`).

Der `else`-Block ist optional. Das heißt, eine `if`-Bedingung funktioniert auch ohne einen `else`-Block. 


````python
zahl = int(input("Gebe eine Zahl ein: "))

if zahl < 100:
  print(f"{zahl} kleiner als 100")
else:
  print(f"{zahl} ist größer gleich 100")
````



---
**Mehrere If-Anweisungen**

*   36,5°C - 38,0°C: Normal-Temperatur
*   38,1°C - 38,5°C: Leichtes Fieber
*   38,6°C - 39,0°C: Mäßiges Fieber
*   39,1°C - 39,9°C: Hohes Fieber
*   40,0°C - 42,0°C: Sehr hohes Fieber


````python
grad = 38.55

if grad <= 38.0 and grad >= 36.5:
  print("Normal-Temperatur")
elif grad >= 38.1 and grad <= 38.5:
  print("Leichtes Fieber")
elif grad >= 38.6 and grad <= 39.0:
  print("Mäßiges Fieber")
elif grad >= 39.1 and grad <= 39.9:
  print("Hohes Fieber")
elif grad >= 40.0:
  print("Sehr hohes Fieber")
else:
  print("Unterkühlt")
````

Über übergeordnete Datenstrukturen iterieren 

````python
matrix = [
  [1, 2, 3],
  [4, 5, 6],
  [7, 8, 9]
]

# Für jede Zeile i in der Matrix ...
for i in matrix:
  print(i)      # ... gebe Zeile i aus und ...
  for j in i:   # ... für jedes Element j der Zeile i ...
    print(j)    # ... gebe das Element j aus.

# Alternative Schreibweise (nur Elemente):
print(*(j for i in matrix for j in i))
````

# `try` und `except`

Mit dem `try`-Block kann man Fehler abfangen, sobald sie auftreten. Bei einem Fehler wird der Inhalt des `except`-Blocks ausgeführt.

````python
try:
  # Versuche diesen Code auszuführen
  alter = int(input("Bitte geben Sie ihr ALter an: "))
  print(alter)
except ValueError:
  # Wenn ein ValueError aufgetreten ist, dann mache folgendes:
  print("Bitte gebe eine Zahl ein!")
except:
  # Wenn irgendein Error aufgetreten ist:
  print("Ein Fehler ist aufgetreten!")
````

# `break`, `continue` und `pass`

* `break`: Die `break`-Anweisung beendet die aktuelle Schleife und setzt die Ausführung nach der Schleife Anweisung fort.
* `continue`: Die `continue`-Anweisung bricht den aktuellen Schleifendurchlauf ab und beginnt den nächsten.
* `pass`: Die `pass`-Anweisung ist ein Platzhalter für leere Codeblöcke in z.B. Schleifen oder Bedingungen.





---

`break`

````python
for i in "Python":
  if i == "t":
    break
  print(i)
````



---


`continue`

````python
for i in "Python":
  if i == "t":
    continue
  print(i)
````



---
`pass`


````python
for i in "Python":
  if i == "t":
    pass
  print(i)
````

Erstelle eine Funktion, die dir errechnet wie viele Münzen du für eine Bestimmte Eingabe von Cents erhältst.



>Beispiel: 200 Cent = eine 2€ Münze

>`Anzahl der Münzen: 1`



>Beispiel: 250 Cent = eine 2€ Münze und eine 50 Cent Münze

>` Anzahl der Münzen: 2`

>Beispiel: 75 Cent = eine 50 Cent Münze, eine 20 Cent Münze, eine 5 Cent Münze                                                                                                                                    
>`Anzahl der Münzen: 3`




````python
def zähleMünzen(anzahl):
  münzen = [200, 100, 50, 20, 10, 5, 1]
  anzahlmünzen = 0

  for münze in münzen:
    anzahlmünzen = anzahlmünzen + anzahl // münze
    anzahl = anzahl % münze

  return anzahlmünzen


cents = int(input("Anzahl der Cents: "))
print(f"Anzahl der Münzen beträgt: {zähleMünzen(cents)}")
````

# Klassen

Python ist eine objekt-orientierte Programmiersprache. Dies bedeutet dass in Pythen nahezu alles ein Objekt ist. Beispiele sind `str`, `int` oder `list` welche Objekt-Typen (= Klasse) sind. Wenn wir ein neues Objekt erzeugen bspw. `name = "Tim"`, dann ist die Variable `name` eine Instanz der Klasse `str`.

Jedes Objekt besteht aus Attributen (Eigenschaften) und Methoden (Funktionen). Für den Objekt-Typ des `dict` kennen wir beispielsweise die Methode `keys()`. 

Python erlaubt es uns jedoch auch eigene Klassen zu definieren, wozu es das Keyword `class` gibt.

````python
# Definition der Klasse
class Student():
  name = "Max Mustermann"
  alter = 21
  kurse = ["Vorkurs Informatik", "Informatik 1", "Mathematik 1"]

# Erzeuge ein Objekt (Instanz) der Klasse
schüler = Student()

# Hinweis: print(*liste) <-> print(liste[0], liste[1], ..., liste[-1])  
print(f"Unser Schüler heist {schüler.name}, ist {schüler.alter} Jahre alt und"
      f" in folgende Kurse eingeschrieben:", *schüler.kurse, sep="\n - ", end="\n\n")
````
